## Excercise 2 : 
Write a program to prompt for a file name, and then read through the file and look for lines of the form:

*`X-DSPAM-Confidence: 0.8475`*

When you encounter a line that starts with *`X-DSPAM-Confidence:`* pull apart the line to extract the floating-point number on the line. Count these lines and then compute the total of the spam confidence values from these lines. When you reach the end of the file, print out the average spam confidence.



Test your file on the mbox.txt and mbox-short.txt files.

```python
print(f"""
Path for mbox-short : ../Text Files/mbox-short.txt
or
Path for mbox : ../Text Files/mbox.txt
-------------------------------------------------------
""")

path = input("Enter the file path")
file = open(path)

count = 0
sum = 0

for line in file:
    if "X-DSPAM-Confidence:" in line:
        count += 1
        x = line.split()
        sum += float(x[1])

file.close()
print("Average spam confidence: ", sum/count)
```

## Excercise 3 : 
Sometimes when programmers get bored or want to have a bit of fun, they add a harmless *`Easter Egg`* to their program. Modify the program that prompts the user for the file name so that it prints a funny message when the user types in the exact file name `na na boo boo`. The program should behave normally for all other files which exist and dont exist. Here is a sample execution of the program:





```python
path = input("Enter the file path")
# path is "../Text Files/mbox-short.txt"
try:
    file = open(path)
    count = 0
    sum = 0

    for line in file:
        if "X-DSPAM-Confidence:" in line:
            count += 1
            x = line.split()
            sum += float(x[1])
    file.close()
    print("Average spam confidence: ", sum/count)

except:
    if path.strip().lower() == "na na boo boo":
        print("NA NA BOO BOO TO YOU - You have been punk'd!")
    else:
        print("File cannot be opened: ", path)
```
